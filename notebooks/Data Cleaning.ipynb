Convert the Lest_X and Lest_Y to a more usable format. Given the ranges, calculates their centers.

```python
def calculate_center_x(row):
    try:
        coords = row['Lest_X'].split("-")
        return (int(coords[0]) + int(coords[1])) / 2
    except:
        return "NaN"
        
def calculate_center_y(row):
    try:
        coords = row['Lest_Y'].split("-")
        return (int(coords[0]) + int(coords[1])) / 2
    except:
        return "NaN"

data["x"] = data.apply(lambda row: calculate_center_x(row),axis=1)
data["y"] = data.apply(lambda row: calculate_center_x(row),axis=1)
```
